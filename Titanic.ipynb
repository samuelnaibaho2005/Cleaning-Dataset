```python
import pandas as pd
```

```python
df = pd.read_csv('tested.csv')
```

menampilkan tabel

```python
df
```

Melihat informasi tabel seperti tipe data dan nilai yang kosong

```python
df.info()
```

Mengecek nilai yang kosong dan menjumlahkannya

```python
df.isnull().sum()
```

Memilih kolom age lalu mengisi nilai yang kosong tersebut dengan nilai median, bisa juga mengisinya dengan mean(rata-rata) tergantung jenis datanya

```python
df['Age'].fillna(df['Age'].median())
```

```python
df.isnull().sum()
```

menghapus kolom Cabin karena banyak nilai yang kosong, sehingga sulit untuk di isi menggunakan mean atau median. menggunakan inplace agar data yang asli digantikan juga

```python
df.dropna(subset='Cabin', inplace=True)
```

jika menggunakan dropna, itu hanya akan menghapus isi dari kolom Cabin, tetapi nama kolom Cabin tetap ada. Dan jika menggunakan drop maka akan menghapus nama kolom dan isinya juga

```python
df.drop(columns=['Cabin'], inplace=True)
```

Mengimport library matplotlib dan seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

untuk membuat ukuran grafik nya (8 x 5 inci)

```python
plt.figure(figsize=(5, 5))
```

```python
sns.histplot(df['Age'], bins=30, kde=True, color='skyblue') #membuat histogram
plt.title('Data Grafik Usia Penumpang Titanic') #judul grafiknya
plt.xlabel('Usia') #untuk sumbu di x
plt.ylabel('Jumlah Penumpang') #data di sumbu y
plt.show() # menampilkan grafik
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Pclass', hue='Survived', palette='coolwarm')
plt.title('Hubungan Kelas Kabin')
plt.xlabel('Kelas Tiket')
plt.ylabel('Jumlah Penumpang')
plt.legend('Meninggal', 'Selamat')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Sex", hue="Survived", palette="viridis")
plt.title("Hubungan Jenis Kelamin dan Kelangsungan Hidup")
plt.xlabel("Jenis Kelamin")
plt.ylabel("Jumlah Penumpang")
plt.legend(["Meninggal", "Selamat"])
plt.show()
```
